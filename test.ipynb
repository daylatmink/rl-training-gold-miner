```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters())
```

```python
state = json_to_state('state.json')
```

```python
agent = Qtention()
```

```python
type, feat, mov_idx, mov_feat = agent.embedder.preprocess_state(state)
type = torch.tensor(type, dtype=torch.long).unsqueeze(0)
feat = torch.tensor(feat, dtype=torch.float32).unsqueeze(0)
mov_idx = torch.tensor(mov_idx, dtype=torch.long).unsqueeze(0)
mov_feat = torch.tensor(mov_feat, dtype=torch.float32).unsqueeze(0)
```

## Test Buffer with Preprocessed Tensors

```python
from train_dqn import ReplayBuffer

# Create buffer
buffer = ReplayBuffer(capacity=10)

# Push a transition (preprocessed)
buffer.push(
    type, feat, mov_idx, mov_feat,  # Current state (đã preprocess ở cell trước)
    action=1,
    reward=100.0,
    next_type_ids=type,  # Next state (giả lập bằng cách reuse)
    next_item_feats=feat,
    next_mov_idx=mov_idx,
    next_mov_feats=mov_feat,
    done=False
)

print(f"Buffer size: {len(buffer)}")
print(f"Each transition has 11 elements: (type_ids, item_feats, mov_idx, mov_feats, action, reward, next_type_ids, next_item_feats, next_mov_idx, next_mov_feats, done)")
```
